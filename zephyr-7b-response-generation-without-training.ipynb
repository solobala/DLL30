```python
def remove_emoji(text: str) -> str:
    """Вспомогательная функция для удаления смайликов из текста
    Args: text: str - исходный текст
    Return: str - текст без смайликов"""
    # Паттерн для поиска emoji
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # эмодзи с улыбками
                               u"\U0001F300-\U0001F5FF"  # эмодзи с надписями и символами
                               u"\U0001F680-\U0001F6FF"  # транспортные эмодзи
                               u"\U0001F1E0-\U0001F1FF"  # флаги стран
                               u"\U00002500-\U00002BEF"  # китайские иероглифы
                               u"\U00002702-\U000027B0"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               u"\U0001f926-\U0001f937"
                               u"\U00010000-\U0010ffff"
                               u"\u200d"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\u3030"
                               u"\ufe0f"
                               "]+", flags=re.UNICODE)

    # Удаление emoji из текста
    return emoji_pattern.sub(r'', text)

df['review'] = df['review'].apply(remove_emoji)
```

## Генерация ответов с использованием промпта, настраиваемого по тексту отзыва, и tokenizer.apply_chat_template() (zero-shot prompting)

```python
def get_messages_(review):
    """
    Функция для генерации промпта по стратегии zero-shot prompting.
    Передается только текст отзыва. Примеры генерации не предоставляются 
    
    Args:
        review - текст отзыва
    """
    messages = [
            {
                "role": "system",
                "content": f"""Вы - русский продавец и получили от покупателя отзыв о товаре. 
                Напишите ответ на русском языке. Не используйте английские слова.
                В ответе выразите понимание проблемы и 
                сообщите о предпринимаемых мерах по ее устранению.
                """
            },
            {
                "role": "user",
                "content": review,
            }]
    return messages
```

```python
# Проверка. Генерируем сообщения для взятого в примере отзыва. 
messages = get_messages_(ref_review)
messages
```

```python
# С помощью метода apply_chat_template применяем к сообщениям шаблон, получим промпт для модели
pipe.tokenizer.apply_chat_template(
            messages, tokenize=False, add_generation_prompt=True
        )
```

```python
def generate_response_(pipe, ref_review, max_new_tokens):
    """Функция генерирует ответ на отзыв review - zero-shot prompting"""
    messages = get_messages_(ref_review)

    prompt = pipe.tokenizer.apply_chat_template(
            messages, tokenize=False, add_generation_prompt=True
        )
    outputs = pipe(
                prompt,
                max_new_tokens=max_new_tokens,
                do_sample=True,
                temperature=0.6,
                top_k=50,
                top_p=0.95,
                pad_token_id=0,
            )

    # выделение ответа из тела сгенерированного сообщения
    generated_response = outputs[0]["generated_text"].split('<|assistant|>')[1]
    
    # Обрезка незаконченного "хвоста'
    if '. ' in generated_response:
        generated_response = '. '.join(generated_response.split('. ')[:-1]) + '.'
    
    return generated_response
```

```python
%%time
# Проверка работы функции на 1 отзыве
resp = generate_response_(pipe, ref_review, max_new_tokens)
resp
```

```python
def save_results(file_path, neg_texts, generated_responses):
    """ Функция для сохранения результатов генерации. Генрация выполняется батчами, 
    результаты сохраняются по мере обработки батчец"""
    if os.path.exists(file_path):
        try:
            # Пытаемся загрузить данные
            with open(file_path, 'rb') as f:
                all_texts, all_responses = pickle.load(f)
        except EOFError:
            # Если файл пуст, инициируем списки
            all_texts, all_responses = [], []
        
        # Добавляем новые данные
        all_texts += neg_texts
        all_responses += generated_responses
        
        # Сохраняем обновленные данные
        with open(file_path, 'wb') as f:
            pickle.dump((all_texts, all_responses), f)              
    else:
        # Сохраняем отзывы и сгенерированные ответы на них
        with open(file_path, 'wb') as f:
            pickle.dump((neg_texts, generated_responses), f)
```

```python
def print_results(neg_texts,generated_responces,neg_responses):
    """Функция для вывода на печать результатов генерации"""
    if neg_responses is not None:

        for i, j, k in zip(neg_texts, generated_responces, neg_responses):
                print('Отзыв: ', i)
                print('-----------------------------------------------')
                print('Сгенерированный ответ: ', j)
                print('-----------------------------------------------')
                print('Референсный ответ: ', k)
                print('-----------------------------------------------')
                print()
    else:
        for i, j in zip(neg_texts, generated_responces):
                print('Отзыв: ', i)
                print('-----------------------------------------------')
                print('Сгенерированный ответ: ', j)
                print('-----------------------------------------------')
                print()
```

```python
def generate_all_(pipe,                 
                 max_new_tokens,
                 all_texts_test,  
                 ref_responses=None, 
                 is_save_responses=False):
    """Функция генерации ответов на список отзывов
    Args:
    max_new_tokens - максимальное к-во новых токенов
    all_text_test - список отзывов для генерации ответов,
    ref_responses - референсные ответы к отзывам, передаваемым для генерации
    is_save_responses - признак, следует ли сохранять сгенерированнеы ответы   
    """
    # Инициализируем список для хранения сгенерированных ответов, 
    generated_responses = []
    
    # Цикл по каждому отзыву
    for review in tqdm(all_texts_test):
        
        # Генерация ответа на отзыв
        generated_response = generate_response_(pipe, 
                                               review,
                                               max_new_tokens)
        generated_responses.append(generated_response)

    if is_save_responses:
        # сохраняем отзывы и сгенерированные ответы в файле
        save_results('/kaggle/working/generated_responses_zephyr_0.pkl', 
                     all_texts_test, generated_responses)
    return generated_responses
```

## Генерация ответов с использованием промпта, настраиваемого по тексту отзыва, tokenizer.apply_chat_template() и стратегии few shot prompting

```python
def get_documents(review,  all_texts_test, test_embeddings, train_embeddings, all_ref_responses_train):
    # Это индекс отзыва review в тестовой подвыборке
    id = all_texts_test.index(review)
    
    # Это эмбеддинг отзыва review из тестовой подвыборки
    emb = test_embeddings[id]
    
    # по косинусной близости между emb и эмбеддингами из train_embeddings находим 5 ближайших соседей
    index = nmslib.init(method='hnsw', space='cosinesimil')
    index.addDataPointBatch(train_embeddings, ids=list(range(len(train_embeddings))))
    index.createIndex(print_progress=False)
    
    # Получаем их индексы и расстояния (distance =1 - cosine)
    ids, distances = index.knnQuery(emb, k=5)
    
    # по найденным индексам отбираем наиболее подходящие варианты ответов в all_ref_responses_train
    # и добавляем их в текст messages
    documents=[]
    for i, d in zip(ids, distances):
        documents.append(all_ref_responses_train[i])
    
    # Создаем строку, содержащую каждый элемент списка documents
    documents_content = "\n".join([doc for doc in documents])
    return(documents_content)
```

```python
def get_messages(review,  all_texts_test, test_embeddings, 
                 train_embeddings, all_ref_responses_train):
    """
    Функция для настройки промпта языковой модели. 
    Настройка происходит путем поиска отзывов, 
    максимально близких к текущему, отбора ответов на эти отзывы и
    передачи ответов в промпт в качестве примеров при генерации
    Args:
        review - текст отзыва
        all_texts_test - общий список отзывов,
        test_embeddings - эмбеддинги отзывов из all_texts_test
        train_embeddings - эмбеддинги отзывов из train выборки
    """
    
    documents_content = get_documents(review,  all_texts_test, 
                                      test_embeddings, train_embeddings, 
                                      all_ref_responses_train)
    messages = [
            {
                "role": "system",
                "content": f"""Вы - русский продавец и получили от покупателя отзыв о товаре. 
                Напишите ответ на русском языке. Не используйте английские слова.
                В ответе выразите понимание проблемы и сообщите 
                о предпринимаемых мерах по ее устранению.
                При подготовке ответа в качестве примеров используйте следующие варианты:
                {documents_content}
                """
            },
            {
                "role": "user",
                "content": review,
            }]
    return messages
```

```python
def generate_response(pipe, ref_review,train_embeddings, 
                      all_texts_test, test_embeddings, max_new_tokens, 
                      all_ref_responses_train):
    """Функция генерирует ответ на отзыв review - few-shot prompting"""
    messages = get_messages(ref_review, all_texts_test, test_embeddings, 
                            train_embeddings, all_ref_responses_train)

    prompt = pipe.tokenizer.apply_chat_template(
            messages, tokenize=False, add_generation_prompt=True
        )
    outputs = pipe(
                prompt,
                max_new_tokens=max_new_tokens,
                do_sample=True,
                temperature=0.6,
                top_k=50,
                top_p=0.95,
                pad_token_id=0,
            )

    # выделение ответа из тела сгенерированного сообщения
    generated_response = outputs[0]["generated_text"].split('<|assistant|>')[1]
    
    # Обрезка незаконченного "хвоста'
    if '. ' in generated_response:
        generated_response = '. '.join(generated_response.split('. ')[:-1]) + '.'
    
    return generated_response
```

```python
def generate_all(pipe,                 
                 max_new_tokens,
                 test_embeddings,
                 train_embeddings,
                 all_texts_test,  
                 ref_responses=None, 
                 is_save_responses=False, 
                 all_text_responses_train=all_ref_responses_train):
    """Функция генерации ответов на на список отзывов - few-shot prompting
    Args:
    max_new_tokens - максимальное к-во новых токенов
    test_embeddings - эмбеддинги для перердаваеых отзывов
    train_embeddings - эмбеддинги для учебных отзывыв, среди которых ищем похожие на тестовые
    all_text_test - список отзывов для генерации ответов,
    ref_responses - референсные ответы к отзывам, передаваемым для генерации
    is_save_responses - признак, следует ли сохранять сгенерированнеы ответы
    all_text_responses_train - список учебных ответов, среди которых выбираются обрпазцы для промпта)
    
        """
    generated_responses = []
    
    # Цикл по каждому отзыву
    for review in tqdm(all_texts_test):
        
        # Генерация ответа на отзыв
        generated_response = generate_response(pipe, 
                                               review,
                                               train_embeddings, 
                                               all_texts_test, 
                                               test_embeddings, 
                                               max_new_tokens, 
                                               all_text_responses_train)
        generated_responses.append(generated_response)

    if is_save_responses:
        # сохраняем отзывы и сгенерированные ответы в файле
        save_results('/kaggle/working/generated_responses_zephyr_1.pkl',
                     all_texts_test, generated_responses)
    
    return generated_responses
```
